## Now we can generate the initial momentum dependence in milne coordinates

NOTE : $$\hat{p}_{\eta} = p_{\eta}$$

We have the following relation between polar and cartesian coordinates:

NOTE that we are restricting our spacetime to midrapididty: $z = \eta = 0$
$$\hat{\phi} = -\sin(\phi)\hat{x} + \cos(\phi)\hat{y}$$
$$p_{\phi} = \vec{p} \cdot \hat{\phi} = -\sin(\phi)p_x + \cos(\phi)p_y =$$

We can use that 
$$p_x = |\vec{p}|v_x = |\vec{p}|\sin(\theta_p)\cos(\phi_p)$$
$$p_y = |\vec{p}|v_y =|\vec{p}|\sin(\theta_p)\sin(\phi_p)$$

Since we have that 
$$t = \tau \cosh(\eta)$$
$$z = \tau \sinh(\eta)$$

if follows from chain rule that $$p_{\eta} = \frac{\partial t}{\partial \eta} p_t + \frac{\partial z}{\partial \eta}p_z = \tau \sinh(\eta)p_t + \tau \cosh(\eta) p_z = zp_t + tp_z = tp_z = |\vec{p}|tv_z$$
where we used that z = 0 in the last equality.

```python
phip_grid = np.linspace(0., 2*np.pi, 20)
p_mag_grid = np.linspace(0., 1., 15) #the magnitude |p|
vz_grid = np.linspace(0., 1., 15)

#because of the rotational symmetry, we can compute the distribution function at some fixed coordinate
#transverse angle phi, and then compute the distribution function as a function of momentum phi_p where phi_p is
#defined w.r.t. phi
#then we only need to compute it as a function of radius, at a fixed angle....
soln_milne_2 = np.zeros( (len(r_vals), len(p_mag_grid), len(phip_grid), len(vz_grid) ) )
time_b = time.time()
for ir, r in enumerate(r_vals):
    #spacetime coordinates in milne 
    phi = 0.
    sin_phi = np.sin(phi)
    cos_phi = np.cos(phi)
    #spacetime coordinates in Gubser
    rho = rho_func(tau0, r, q0)
    theta = theta_func(tau0, r, q0)    
    for ivz, vz in enumerate(vz_grid):
        sin_theta_p = np.sqrt(1.-vz**2)
        cos_theta_p = vz
        for iphip, phip in enumerate(phip_grid):
            vx = cos_theta_p * np.cos(phip)
            vy = cos_theta_p * np.sin(phip)    
            for ip_mag, p_mag in enumerate(p_mag_grid):
                #use mass shell condition p_mu p^mu = 0
                p_T = p_mag * sin_theta_p # p_T = |p| sin \theta_p
                p_eta = p_mag * tau0 * vz
                p_tau = np.sqrt( p_T**2 + (p_eta**2/tau0**2) )
                p_phi = p_mag * (-sin_phi * vx + cos_phi * vy) 
                
                p_theta_hat = p_theta_func(rho, theta, q0, p_tau, p_T)
                p_omega_hat = p_omega_hat_func(p_theta_hat, p_phi, theta)    
                #check that arguments lie inside bounds
                p_omega_hat = np.minimum(p_omega_hat, p_omega_hat_grid[-1])    
                coords = (rho, theta)
                #print("(rho, theta) = " + str(coords))
                arg = (rho, p_omega_hat, p_eta)
                #print("(rho, p_omega_hat, p_eta) = " + str(arg))
                soln_milne_2[ir, ip_mag, iphip, ivz] = f_soln_interp(arg)
time_a = time.time()
time_diff = time_a - time_b
print("Took " + str(time_diff) + "sec")
```
